Load hmda clean, then get # of loans for all banks, then get list of banks between 50% and 200% of BOW

```python
# find the unique banks, convert lei to bank names, need another table to merge
# add bank names to corresponding to lei, 
# sum by the number of loans for each bank
```

```python
import os
import csv
import pandas as pd
import numpy as np

os.makedirs('../input_data_clean/',exist_ok=True)
```

```python
hmda = pd.read_csv('../input_data_clean/state_AZ-CA_clean_MINI.csv')
hmda = hmda.replace(["Exempt","1111","8888","9999","NA"], np.nan)  
```

```python
#filter valid loan/application by only including action_taken = 1, 2, or 3
hmda = hmda[(hmda['action_taken'] == 1) | (hmda['action_taken'] == 2) | (hmda['action_taken'] == 3)]
```

```python
groupedlei = hmda.groupby('lei').size().reset_index(name='Loan Count')
groupedlei
```

```python
lei_list = pd.read_csv('../input_data/20230214-0000-gleif-goldencopy-lei2-golden-copy.csv',low_memory=False)
```

```python
lei_list.rename(columns={'LEI': 'lei'}, inplace=True)
```

```python
lei_list.sample(50)
```

```python
lei_merged_df = pd.merge(groupedlei, lei_list[['lei', 'Entity.LegalName']], on='lei', how='inner')
```

```python
lei_merged_df.rename(columns={'Entity.LegalName': 'Bank Name'}, inplace=True)
lei_merged_df = lei_merged_df.reindex(columns=['lei','Bank Name', 'Loan Count'])
lei_merged_df
```

```python
BOW_loan_count = lei_merged_df[lei_merged_df['lei']=='QOT5WN9RBKQTFRVKEV31'].iloc[0]['Loan Count']
```

```python
print("BOW's #Loan:", BOW_loan_count)
```

```python
BOW_low_benchmart = BOW_loan_count*0.5
BOW_high_benchmart = BOW_loan_count*2
print(BOW_low_benchmart)
print(BOW_high_benchmart)
```

```python
filter_banklist = lei_merged_df[(lei_merged_df['Loan Count'] >= BOW_low_benchmart ) & (lei_merged_df['Loan Count']<= BOW_high_benchmart)]
```
